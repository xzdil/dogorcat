```python
from typing import Any, Dict, List, Optional, Sequence, cast

from llama_index.core.async_utils import DEFAULT_NUM_WORKERS, run_jobs
from llama_index.core.bridge.pydantic import Field, PrivateAttr
from llama_index.core.extractors.interface import BaseExtractor
from llama_index.core.llms.llm import LLM
from llama_index.core.prompts import PromptTemplate
from llama_index.core.schema import BaseNode, TextNode
from llama_index.core.service_context_elements.llm_predictor import (
    LLMPredictorType,
)
from llama_index.core.settings import Settings
from llama_index.core.types import BasePydanticProgram


DEFAULT_TITLE_NODE_TEMPLATE = """\
Контекст: {context_str}. Дайте заголовок, который суммирует все \
уникальные объекты, названия или темы, найденные в контексте. Заголовок: """


DEFAULT_TITLE_COMBINE_TEMPLATE = """\
{context_str}. На основании вышеуказанных названий и содержания кандидатов, \
каково полное название этого документа? Заголовок: """


class CustomTitleExtractor(TitleExtractor):
    is_text_node_only: bool = False  # can work for mixture of text and non-text nodes
    llm: LLMPredictorType = Field(description="LLM, который будет использоваться для генерации.")
    nodes: int = Field(
        default=5,
        description="Количество узлов, из которых нужно извлечь заголовки.",
        gt=0,
    )
    node_template: str = Field(
        default=DEFAULT_TITLE_NODE_TEMPLATE,
        description="Шаблон промпта для извлечения заголовков.",
    )
    combine_template: str = Field(
        default=DEFAULT_TITLE_COMBINE_TEMPLATE,
        description="Шаблон промпта для объединения заголовков.",
    )

    def __init__(
        self,
        llm: Optional[LLM] = None,
        # TODO: llm_predictor arg is deprecated
        llm_predictor: Optional[LLMPredictorType] = None,
        nodes: int = 5,
        node_template: str = DEFAULT_TITLE_NODE_TEMPLATE,
        combine_template: str = DEFAULT_TITLE_COMBINE_TEMPLATE,
        num_workers: int = DEFAULT_NUM_WORKERS,
        **kwargs: Any,
    ) -> None:
        """Init params."""
        if nodes < 1:
            raise ValueError("num_nodes must be >= 1")

        super().__init__(
            llm=llm or llm_predictor or Settings.llm,
            nodes=nodes,
            node_template=node_template,
            combine_template=combine_template,
            num_workers=num_workers,
            **kwargs,
        )
```

```python
DEFAULT_QUESTION_GEN_TMPL = """\
Вот контекст:
{context_str}

Учитывая контекстную информацию, \
сгенерировать {num_questions} вопросов, которые может предоставить этот контекст \
конкретные ответы на которые вряд ли можно найти где-либо еще.

Могут быть предоставлены краткие описания окружающего контекста \
также. Попробуйте использовать эти сводки, чтобы сформулировать более качественные вопросы \
что этот контекст может ответить.

"""


class CustomQuestionsAnsweredExtractor(QuestionsAnsweredExtractor):
    llm: LLMPredictorType = Field(description="LLM, который будет использоваться для генерации.")
    questions: int = Field(
        default=5,
        description="Количество вопросов, которые нужно сгенерировать.",
        gt=0,
    )
    prompt_template: str = Field(
        default=DEFAULT_QUESTION_GEN_TMPL,
        description="Шаблон промпта, который можно использовать при создании вопросов.",
    )
    embedding_only: bool = Field(
        default=True, description="Использовать ли метаданные только для эмбеддинга."
    )

    def __init__(
        self,
        llm: Optional[LLM] = None,
        # TODO: llm_predictor arg is deprecated
        llm_predictor: Optional[LLMPredictorType] = None,
        questions: int = 5,
        prompt_template: str = DEFAULT_QUESTION_GEN_TMPL,
        embedding_only: bool = True,
        num_workers: int = DEFAULT_NUM_WORKERS,
        **kwargs: Any,
    ) -> None:
        """Init params."""
        if questions < 1:
            raise ValueError("questions must be >= 1")

        super().__init__(
            llm=llm or llm_predictor or Settings.llm,
            questions=questions,
            prompt_template=prompt_template,
            embedding_only=embedding_only,
            num_workers=num_workers,
            **kwargs,
        )
```

```python
DEFAULT_SUMMARY_EXTRACT_TEMPLATE = """\
Вот содержимое раздела:
{context_str}

Кратко изложите основные темы и сущности раздела. \

Краткое содержание: """


class CustomSummaryExtractor(SummaryExtractor):
    llm: LLMPredictorType = Field(description="LLM, который будет использоваться для генерации")
    summaries: List[str] = Field(
        description="Список суммаризаций для извлечения: «self», «prev», «next»."
    )
    prompt_template: str = Field(
        default=DEFAULT_SUMMARY_EXTRACT_TEMPLATE,
        description="Шаблон для использования при создании суммаризаций.",
    )

    _self_summary: bool = PrivateAttr()
    _prev_summary: bool = PrivateAttr()
    _next_summary: bool = PrivateAttr()

    def __init__(
        self,
        llm: Optional[LLM] = None,
        # TODO: llm_predictor arg is deprecated
        llm_predictor: Optional[LLMPredictorType] = None,
        summaries: List[str] = ["self"],
        prompt_template: str = DEFAULT_SUMMARY_EXTRACT_TEMPLATE,
        num_workers: int = DEFAULT_NUM_WORKERS,
        **kwargs: Any,
    ):
        # validation
        if not all(s in ["self", "prev", "next"] for s in summaries):
            raise ValueError("summaries must be one of ['self', 'prev', 'next']")
        self._self_summary = "self" in summaries
        self._prev_summary = "prev" in summaries
        self._next_summary = "next" in summaries

        super().__init__(
            llm=llm or llm_predictor or Settings.llm,
            summaries=summaries,
            prompt_template=prompt_template,
            num_workers=num_workers,
            **kwargs,
        )
```

```python

class CustomKeywordExtractor(KeywordExtractor):

    llm: LLMPredictorType = Field(description="LLM, который будет использоваться для генерации")
    keywords: int = Field(
        default=5, description="Количество ключевых слов для извлечения.", gt=0
    )

    def __init__(
        self,
        llm: Optional[LLM] = None,
        # TODO: llm_predictor arg is deprecated
        llm_predictor: Optional[LLMPredictorType] = None,
        keywords: int = 5,
        num_workers: int = DEFAULT_NUM_WORKERS,
        **kwargs: Any,
    ) -> None:
        """Init params."""
        if keywords < 1:
            raise ValueError("num_keywords must be >= 1")

        super().__init__(
            llm=llm or llm_predictor or Settings.llm,
            keywords=keywords,
            num_workers=num_workers,
            **kwargs,
        )
```

```python
from llama_index.extractors.entity import EntityExtractor
```

```python

from typing import Any, Callable, Dict, Iterable, List, Optional, Sequence

from llama_index.core.bridge.pydantic import Field, PrivateAttr
from llama_index.core.extractors.interface import BaseExtractor
from llama_index.core.schema import BaseNode
from llama_index.core.utils import get_tqdm_iterable
from nltk.tokenize import word_tokenize
from span_marker import SpanMarkerModel

DEFAULT_ENTITY_MAP = {
    "PER": "люди",
    "ORG": "организации",
    "LOC": "локации",
    "ANIM": "животные",
    "BIO": "биологические",
    "CEL": "небесные",
    "DIS": "расстройства",
    "EVE": "собыття",
    "FOOD": "еда",
    "INST": "инструменты",
    "MEDIA": "медиа",
    "PLANT": "растения",
    "MYTH": "мифология",
    "TIME": "времена",
    "VEHI": "машины",
}

DEFAULT_ENTITY_MODEL = "lxyuan/span-marker-bert-base-multilingual-cased-multinerd"

class CustomEntityExtractor(EntityExtractor):
    model_name: str = Field(
        default=DEFAULT_ENTITY_MODEL,
        description="Имя используемой модели SpanMarker.",
    )
    prediction_threshold: float = Field(
        default=0.5,
        description="Порог доверия для принятия прогнозов",
        gte=0.0,
        lte=1.0,
    )
    span_joiner: str = Field(
        default=" ", description="Разделитель между именами сущностей."
    )
    label_entities: bool = Field(
        default=False, description="Включить или нет метки классов сущностей."
    )
    device: Optional[str] = Field(
        default=None, description="Устройство для запуска модели, например «cuda», «cpu»."
    )
    entity_map: Dict[str, str] = Field(
        default_factory=dict,
        description="Сопоставление имен классов сущностей с пригодными для использования именами.",
    )

    _tokenizer: Callable = PrivateAttr()
    _model: Any = PrivateAttr()

    def __init__(
        self,
        model_name: str = DEFAULT_ENTITY_MODEL,
        prediction_threshold: float = 0.5,
        span_joiner: str = " ",
        label_entities: bool = False,
        device: Optional[str] = None,
        entity_map: Optional[Dict[str, str]] = None,
        tokenizer: Optional[Callable[[str], List[str]]] = None,
        **kwargs: Any,
    ):
        self._model = SpanMarkerModel.from_pretrained(model_name)
        if device is not None:
            self._model = self._model.to(device)

        self._tokenizer = tokenizer or word_tokenize

        base_entity_map = DEFAULT_ENTITY_MAP
        if entity_map is not None:
            base_entity_map.update(entity_map)

        super().__init__(
                model_name=model_name,
                prediction_threshold=prediction_threshold,
                span_joiner=span_joiner,
                label_entities=label_entities,
                device=device,
                entity_map=base_entity_map,
                **kwargs,
            )
```

```python
from llama_index.core.node_parser import SentenceSplitter

transformations = [
    SentenceSplitter(),
    CustomTitleExtractor(nodes=5),
    CustomQuestionsAnsweredExtractor(questions=3),
    CustomSummaryExtractor(summaries=["prev", "self"]),
    CustomKeywordExtractor(keywords=10),
    CustomEntityExtractor(prediction_threshold=0.5),
]
```

```python
from llama_index.core.ingestion import IngestionPipeline

pipeline = IngestionPipeline(transformations=transformations)

nodes = pipeline.run(documents=documents)
```

```python
from llama_index.core.schema import TextNode, NodeRelationship, RelatedNodeInfo
import json

def create_json_from_documents(documents):
    data = []
    for doc in documents:
        if "entities" not in doc.metadata:
          doc.metadata["entities"] = []
        doc_dict = {
            "id": doc.id_,
            "embedding": doc.embedding,
            "metadata": {
                "file_name": doc.metadata["file_name"],
                "file_path": doc.metadata["file_path"],
                "file_type": doc.metadata["file_type"],
                "file_size": doc.metadata["file_size"],
                "creation_date": doc.metadata["creation_date"],
                "last_modified_date": doc.metadata["last_modified_date"],
                "document_title": doc.metadata["document_title"],
                "questions_this_excerpt_can_answer": doc.metadata["questions_this_excerpt_can_answer"],
                "section_summary": doc.metadata["section_summary"],
                "excerpt_keywords": doc.metadata["excerpt_keywords"],
                "entities": doc.metadata["entities"]
            },
            "relationships": {}
        }
        if NodeRelationship.SOURCE in doc.relationships:
            doc_dict["relationships"]["source"] = {
                "node_id": doc.relationships[NodeRelationship.SOURCE].node_id,
                "node_type": doc.relationships[NodeRelationship.SOURCE].node_type,
                "metadata": doc.relationships[NodeRelationship.SOURCE].metadata,
                "hash": doc.relationships[NodeRelationship.SOURCE].hash
            }

        if NodeRelationship.NEXT in doc.relationships:
            doc_dict["relationships"]["next"] = {
                "node_id": doc.relationships[NodeRelationship.NEXT].node_id,
                "node_type": doc.relationships[NodeRelationship.NEXT].node_type,
                "metadata": doc.relationships[NodeRelationship.NEXT].metadata,
                "hash": doc.relationships[NodeRelationship.NEXT].hash
            }

        doc_dict["text"] = doc.text
        data.append(doc_dict)

    return json.dumps(data, indent=2), data
```

```python
json_nodes, data_nodes = create_json_from_documents(nodes)
```
